
# Challenger O-Ring Failure Analysis

This notebook analyzes data related to the Challenger Space Shuttle disaster. 
The focus is on understanding the relationship between launch temperature and 
the likelihood of O-ring failures, which played a critical role in the tragic 
event of January 28, 1986.

Key steps include:
- Exploratory Data Analysis (EDA) to understand the dataset.
- Logistic regression to quantify the impact of temperature on failure risk.
- Conclusion with actionable insights.

![Challenger Space Shuttle](challenger_photo.jpeg "Challenger at Launch")

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('o-ring-erosion-only.data',sep='\\s+',header=None)
blowby_df = pd.read_csv('o-ring-erosion-or-blowby.data', header=None)

with open('o-ring-erosion.names') as f:
    description = [line.strip() for line in f if line.strip()]
    
cols = description[-6:-1]
df.columns = cols
df.info()
```


## Exploratory Data Analysis (EDA)

Exploring the dataset to uncover patterns and understand 
key variables. This includes visualizing distributions and relationships 
between temperature, O-ring failures, and other features.


```python
df = df.rename(columns={
    '1. Number of O-rings at risk on a given flight': "At-risk O-rings",
    '2. Number experiencing thermal distress' : 'O-ring Failures',
    '3. Launch temperature (degrees F)': 'Launch Temperature',
    '4. Leak-check pressure (psi)': 'Leak-check Pressure',
    '5. Temporal order of flight':'Flight Number'
})

#add binary failure column
df['Failure Status'] = (df['O-ring Failures'] > 0).astype(int)

df.head()
```

```python
sns.set(style="whitegrid")
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

#distribution of leak-check pressure
sns.histplot(df['Launch Temperature'], kde=True, ax=ax[0], color='purple', bins=10)
ax[0].set_title('Distribution of Launch Temperature', fontsize=14)
ax[0].set_xlabel('Launch Temperature (F)', fontsize=12)
ax[0].set_ylabel('Frequency', fontsize=12)

#leak-check pressure vs launch temp
sns.scatterplot(data=df, x='Launch Temperature', y='O-ring Failures', hue='O-ring Failures', ax=ax[1])
ax[1].set_title('Launch Temperature vs Number of O-Ring Failures', fontsize=14)
ax[1].set_xlabel('Launch Temperature (F)', fontsize=12)
ax[1].set_ylabel('Number of O-Ring Failures', fontsize=12)

#leak-check pressure vs failure
sns.boxplot(data=df, x='Failure Status', y='Launch Temperature',ax=ax[2])
ax[2].set_title('Launch Temperature vs. Failure', fontsize=14)
ax[2].set_xlabel('O-ring Failures (0 = Success, 1 = Failure)', fontsize=12)
ax[2].set_ylabel('Launch Temperature (F)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

#distribution of leak-check pressure
sns.histplot(df['Leak-check Pressure'], kde=True, ax=ax[0], color='purple', bins=10)
ax[0].set_title('Distribution of Leak-Check Pressure', fontsize=14)
ax[0].set_xlabel('Leak-check Pressure (psi)', fontsize=12)
ax[0].set_ylabel('Frequency', fontsize=12)

#leak-check pressure vs launch temp
sns.scatterplot(data=df, x='Launch Temperature', y='Leak-check Pressure', ax=ax[1], color='green')
ax[1].set_title('Launch Temp vs. Launch Temperature', fontsize=14)
ax[1].set_xlabel('Launch Temperature (F)', fontsize=12)
ax[1].set_ylabel('Leak-check Pressure (psi)', fontsize=12)

#leak-check pressure vs failure
sns.swarmplot(data=df, x='O-ring Failures', y='Leak-check Pressure', ax=ax[2], size=7)
ax[2].set_title('Leak-check Pressure vs. O-ring Failures', fontsize=14)
ax[2].set_xlabel('O-ring Failures (0 = Success, 1 = Failure)', fontsize=12)
ax[2].set_ylabel('Leak-check Pressure (psi)', fontsize=12)

plt.tight_layout()
plt.show()
```


## Logistic Regression Analysis

A logistic regression model is used to evaluate the probability of O-ring 
failure based on launch temperature. This approach is appropriate for 
binary outcomes (failure/no failure).

The following steps are performed:
1. Transforming data into a binary failure indicator.
2. Fitting the logistic regression model.
3. Visualizing the relationship between temperature and failure probability.
4. Predicting failure risk at specific launch temperatures.


```python
#add binary failure column
df['Failure Status'] = (df['O-ring Failures'] > 0).astype(int)
```

```python
#fit logistic regression
from sklearn.linear_model import LogisticRegression

X = df['Launch Temperature'].values.reshape(-1, 1)
y = df['Failure Status']

model = LogisticRegression()
model.fit(X, y)

temp_range = np.linspace(df['Launch Temperature'].min(), df['Launch Temperature'].max(), 200).reshape(-1,1)
predicted_probabilities = model.predict_proba(temp_range)[:,1]
```

```python
# Plot the data and logistic regression curve
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Launch Temperature", y="Failure Status", label="Observed Data", color="blue", s=100)
plt.plot(temp_range, predicted_probabilities, label="Logistic Regression Curve", color="red", linewidth=2)

# Customize the plot
plt.title("Logistic Regression: Probability of O-Ring Failure vs. Temperature", fontsize=14)
plt.xlabel("Temperature (F)", fontsize=12)
plt.ylabel("Probability of Failure", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
launchday_temp = [[31]]

#Predict the probability of failure on launch day
predicted_prob = model.predict_proba(launchday_temp)[:, 1]

print(f"Predicted probability of failure at 31F: {predicted_prob[0]:.9f}")
```


## Conclusion

The analysis highlights that lower launch temperatures significantly increase 
the risk of O-ring failure. This finding aligns with the historical context of 
the Challenger disaster, where the launch occurred at an unusually low 
temperature of 31F. 

Armed with this information, one might advocate for postponing the launch until temperatures reach a certain point. Furture work may explore temperature forecasts throughout the launch day, predicting the risk of O-ring failures at various times.
